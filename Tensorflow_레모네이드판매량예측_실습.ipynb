```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```
# 데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
데이터 = pd.read_csv(파일경로) # 판다스로 파일을 읽어들이기
데이터.head()
```

```
# 종속변수, 독립변수
독립 = 데이터[['온도']]
종속 = 데이터[['판매량']]
print(독립.shape, 종속.shape)
```

```
# 모델을 만듭니다.
X = tf.keras.layers.Input(shape=[1]) # 독립변수 칼럼의 수가 1이므로
Y = tf.keras.layers.Dense(1)(X) # 종속변수의 칼럼의 수가 1이므로
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# 모델을 학습합니다.
model.fit(독립, 종속, epochs=10000, verbose=0) # loss 값이 0에 가까워야 함, verbose는 출력을 안해줌
```

```
# 모델을 이용합니다.
model.predict(독립)
```
